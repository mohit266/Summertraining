```
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```
img=cv2.imread('a.jpg')
```

```
# to split into BGR
x,y,z=cv2.split(img) 
myimg=np.zeros((512,512))
```

```
 #             (x1,y1) (x2,y2)    color
cv2.line(myimg,(0,0),(300,300),(300,300,255),3)
cv2.rectangle(myimg,(150,150),(300,300),(255,0,0),-1)
cv2_imshow(myimg)
```

```
cv2_imshow(x)
cv2_imshow(y)
cv2_imshow(z)
```
